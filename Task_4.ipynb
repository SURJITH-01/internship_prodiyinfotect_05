```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sales-Export_2019-2020.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Removed: df["order_value_EUR"] = df["order_value_EUR"].str.replace(",", "").astype(float) - 'order_value_EUR' is already a float

# Removed: df["cost"] = df["cost"].str.replace(",", "").astype(float) as 'cost' is already a float
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
print("Total Sales:", df["order_value_EUR"].sum())
```

```python
print("Average Sales:", df["order_value_EUR"].mean())
```

```python
print("Maximum Sale:", df["order_value_EUR"].max())
```

```python
country_sales = df.groupby("country")["order_value_EUR"].sum()
print(country_sales)
```

```python
category_sales = df.groupby("category")["order_value_EUR"].sum()
print(category_sales)
```

```python
country_sales.plot(kind="bar", figsize=(10,5))
plt.title("Sales by Country")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()
```

```python
df["device_type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Device Type Distribution")
plt.show()
```

```python
daily_sales = df.groupby("date")["order_value_EUR"].sum()

daily_sales.plot(figsize=(12,5))
plt.title("Daily Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
df.to_csv("Cleaned_Sales_Data.csv", index=False)
```
